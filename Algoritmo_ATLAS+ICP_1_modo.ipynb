```python
# n: number of modes
# alpha: list with the desviation from each mode
def create_heart_ATLAS(alpha, pc):
  #n = len(alpha)
  # read H5 file
  #pc = h5.File('UKBRVLV.h5', 'r')
  # Create the n principal modes with alpha[i] times the standard deviation each
  S = np.transpose(pc['MU']) + alpha*np.sqrt(pc['LATENT'][0,0])*pc['COEFF'][0,:]
  # get ED & ES points, & convert to 3 columns matrix [x, y, z]
  N = S.shape[1] // 2
  ed = np.reshape(S[0,:N], (-1,3))
  return np.array(ed)

# loss function
def loss(alpha, data, pc):
  # ATLAS model from alpha parameters
  # print(len(create_heart_ATLAS(alpha, pc)._value))
  alpha = alpha._value if not isinstance(alpha, np.float64) else alpha
  a = create_heart_ATLAS(alpha, pc)
  a = a._value if not isinstance(a, np.ndarray) else a
  model = pv.PolyData(a)

  # find closest points from model to aligned data
  _, closest_points = model.find_closest_cell(data.points, return_closest_point=True)
  # compute error between aligned data and model
  return sum((np.linalg.norm(closest_points-data.points, axis=1))**2)


# algorithm that optimize parameters for PCA from data for the first mode
def ATLAS_ICP(data,pc, N_iter=5, lr=0.1):
  # start with mean shape of the PCA
  alpha_ = 0.0
  print('Valor de alpha inicial: ' + str(alpha_))
  loss_rec = []
  for i in range(N_iter):
    grad = gradloss(alpha_, data, pc)
    print('El gradiente en la iteración {} es {}'.format(i,grad))
    alpha_ = alpha_ - lr * grad
    loss_rec.append(loss(alpha_, data, pc))
  return alpha_, loss_rec
```

```python
from scipy.spatial import cKDTree
```

```python
def create_heart_ATLAS(alpha, pc):
  #n = len(alpha)
  # read H5 file
  #pc = h5.File('UKBRVLV.h5', 'r')
  # Create the n principal modes with alpha[i] times the standard deviation each
  S = np.transpose(pc['MU']) + alpha*np.sqrt(pc['LATENT'][0,0])*pc['COEFF'][0,:]
  # get ED & ES points, & convert to 3 columns matrix [x, y, z]
  N = S.shape[1] // 2
  ed = np.reshape(S[0,:N], (-1,3))
  return np.array(ed)

def loss(alpha, data_points, pc):
  X_points = create_heart_ATLAS(alpha, pc)
  tree = cKDTree(X_points)
  idx_cp = tree.query(data_points)[1]
  return 0.5*np.mean(np.sum((X_points[idx_cp]- data_points)**2, axis = 1))

def gradloss(alpha, data_points, pc):
  X_points = create_heart_ATLAS(alpha, pc)
  tree = cKDTree(X_points)
  idx_cp = tree.query(data_points)[1]
  n = data_points.shape[0]
  valp = pc['LATENT'][0,0]
  vecp = pc['COEFF'][0,:]
  N = vecp.shape[0] // 2
  vecp = np.reshape(vecp[0:N], (-1,3))
  return (1/n)*np.dot((X_points[idx_cp] - data_points).flatten(), (vecp[idx_cp]).flatten()*np.sqrt(valp))

# algorithm that optimize parameters for PCA from data for the first mode
def ATLAS_ICP(data_points, pc, N_iter=5, lr=0.05):
  # start with mean shape of the PCA
  alpha_ = 1.0
  print('Valor de alpha inicial: ' + str(alpha_))
  loss_rec = []
  for i in range(N_iter):
    grad = gradloss(alpha_, data_points, pc)
    alpha_ = alpha_ - lr * grad
    loss_ = loss(alpha_, data_points, pc)
    print('El gradiente en la iteración {} es {}'.format(i,grad))
    loss_rec.append(loss_)
  print('Valor de alpha final: ' + str(alpha_))
  print('Registro de la perdida en cada iteración:' + str(loss_rec))
  return alpha_, loss_rec
```
